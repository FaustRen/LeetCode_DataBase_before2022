- function for leet code problem(Chinese)

```python

def getLeetCodeContent(each_question):
    """
    each_question: The numb of target question:  "175. Combine Two Tables"
    response: Leetcode problem content from OpenAI(text-davinci-003)
    """
    # prompt = "請給我LeetCode {question number}. {question title}的題目，包含：Table、Example、Output、Explanation的內容"
    prompt = f"請給我LeetCode {each_question}的題目，包含：Input的所有Table、Output的Table、以及每個Example裡Input及Output的Table、Explanation的內容，總之100%還原leetcode題目內容"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=2000,
        # n=1,
        # stop=None,
        top_p=1.0,
        temperature=0.1,
    )
    return response


def getLeetCodeSolution(each_question):
    """
    each_question: The numb of target question:  "175. Combine Two Tables"
    response: Leetcode solution from OpenAI(text-davinci-003)
    """
    # prompt = "請給我LeetCode {question number}. {question title}的題目，包含：Table、Example、Output、Explanation的內容"
    prompt = f"請給我LeetCode {each_question}的解答，要能夠在LeetCode上精準通過的MySQL程式碼，我只要程式碼，其他不要"
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=1000,
        # n=1,
        # stop=None,
        top_p=1.0,
        temperature=0.5,
    )
    return response
```

# Test.2 use loop to get each LeetCode Problem's content and solution.

```python
# use three problems for testing.
leetcode_all_database_2021=["175. Combine Two Tables",
                            "176. Second Highest Salary",
                            "177. Nth Highest Salary",
                            ]
```

```python
for each_question in leetcode_all_database_2021:
    # Problem content & Solution
    problem_content=getLeetCodeContent(each_question)
    # print(problem_content.choices[0].text)


    problem_solution=getLeetCodeSolution(each_question)
    # print(problem_solution.choices[0].text)
    
    # content & solution
    content=problem_content.choices[0].text+problem_solution.choices[0].text
    print(content)
    # save to txt
    leetcode_database_problem="LeetCode_DataBase_"+each_question+".txt"
    with open(leetcode_database_problem, "w") as f:
        f.write(content)
    # print(each_question)

    
```
